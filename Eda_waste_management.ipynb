#Exploratory Data Analysis(EDA)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Waste-Management-Data-DNCC.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

note: No Missing Values found

Data Distributions

```python
# Plot histograms for numerical columns
data.hist(figsize=(15, 12))
plt.tight_layout()
plt.show()
plt.tight_layout()

```

```python
# Boxplot for identifying outliers
plt.figure(figsize=(8, 4))
sns.boxplot(data=data.select_dtypes(include=['float64', 'int64']))
plt.xticks(rotation=90)
plt.show()
```

Label Encoding

```python

# Check columns with categorical variables
categorical_columns = data.select_dtypes(include=['object', 'category']).columns

print("Categorical columns:", categorical_columns)

```

```python
data['Condition of waste Collection vehicles'].unique()
```

```python
data['Condition of waste Collection vehicles'].replace(['Good','Bad'],[1,0],inplace=True)
```

```python
data['Availaility of waste education program'].unique()
```

```python
data['Availaility of waste education program'].replace(['YES', 'Yes', 'No'],[1,1,0],inplace=True)
```

```python
data['Public satisfaction with waste management services'].unique()
```

```python
data['Public satisfaction with waste management services'].replace(['Satisfied', 'Not Satisfied'],[1,0],inplace=True)
```

```python
data['Type of waste'].unique()
```

```python
# Check columns with categorical variables
categorical_columns = data.select_dtypes(include=['object', 'category']).columns

print("Categorical columns:", categorical_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
le = LabelEncoder()

# Apply Label Encoding to each categorical column
for col in categorical_columns:
    data[col] = le.fit_transform(data[col])

data.head()
```

Correlation Matrix

```python
# Compute the correlation matrix
corr_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.1f')
plt.title("Correlation Matrix")
plt.show()
```
