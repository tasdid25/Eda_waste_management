```python
from sklearn.preprocessing import LabelEncoder

# Initialize the label encoder
le = LabelEncoder()

# Apply Label Encoding to each categorical column starting from the second column
for col in categorical_columns[1:]:
    data[col] = le.fit_transform(data[col])

# Display the first few rows of the modified DataFrame
data.head()

```

```python

#sample plot
data.plot(x='Month', y='Population Density',figsize=(4,3))
```

```python
# Number of subplots required
n = len(data.columns) - 1  # Number of columns excluding 'Month'
nrows = (n + 1) // 2  # Number of rows needed for 2 columns
# Reduce figsize by 20% (multiply by 0.8)
fig, axes = plt.subplots(nrows=nrows, ncols=2, figsize=(16, nrows * 4))

# Flatten axes array if necessary
axes = axes.flatten()

# Plot each column in a separate subplot
for i, column in enumerate(data.columns[1:]):
    ax = axes[i]
    # Check if the column contains numeric data
    if pd.api.types.is_numeric_dtype(data[column]):
        data.plot(x='Month', y=column, ax=ax, marker='o', linestyle='-')
        ax.set_title(f'Plot of {column}')
        ax.set_xlabel('')
        #ax.set_ylabel(column)
        ax.grid(True)
    else:
        ax.set_visible(False)  # Hide subplot for non-numeric data
        print(f"Skipping {column} - Non-numeric data")

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)

# Add padding between subplots
plt.subplots_adjust(hspace=2.3, wspace=0.5)  # hspace for vertical, wspace for horizontal

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Month' column to datetime format
data['Month'] = pd.to_datetime(data['Month'], format='%Y-%m')

# Extract the year from the 'Month' column
data['Year'] = data['Month'].dt.year

# Plot frequency of entries for each year
yearly_counts = data['Year'].value_counts().sort_index()  # Count frequency of each year

# Set figure size
plt.figure(figsize=(4, 3))

# Plot the bar chart
ax = yearly_counts.plot(kind='bar', color='skyblue')

# Add title with padding
plt.title('Frequency of Entries per Year', pad=20)  # Adding 20 pixels of padding below the title

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.grid(True)

# Annotate the bars with the values
for i in ax.containers:
    ax.bar_label(i, label_type='edge')  # Label the bars

# Show the plot
plt.show()
```
